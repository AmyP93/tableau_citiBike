```python
import pandas as pd
import glob

all_files = glob.glob("cleaned/*.csv")

li = []

for filename in all_files:
    li.append(pd.read_csv(filename, index_col=None, header=0))

frame = li[0].append(li[1::], ignore_index=True)
```
