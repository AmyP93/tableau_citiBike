```python
import pandas as pd
import glob
import datetime as dt

all_files = glob.glob("cleaned/*.csv")

li = []

for filename in all_files:
    li.append(pd.read_csv(filename, index_col=None, header=0))

frame = li[0].append(li[1::], ignore_index=True)
```

```python
#split date column into date and time
    
#change time to hours/minutes
def convert_time(date):

    try: 
        
        if len(str(date)) >4:   
            date = str(date)
            hour=dt.datetime.strptime(date[7::],'%H:%M:%S').hour*60
            minute=dt.datetime.strptime(date[7::],'%H:%M:%S').minute
            return int(hour + minute)
    
        else:
            return int(round(int(date)/60, 0))
        
    except:
        return 0
    
```

```python
frame["tripduration"] = frame["tripduration"].map(lambda x: convert_time(x))
```
